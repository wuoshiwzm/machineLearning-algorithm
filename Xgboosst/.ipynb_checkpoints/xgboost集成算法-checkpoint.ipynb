So random forests and boosted trees are not different in terms of model, the difference is how we train them

This means if you write a predictive service of tree ensembles, you only need to write one of them and they should directly work for both random forests and boosted trees

Instead, we use an additive strategy: fix what we have learned, add one $new$ $tree$ at a time. 

It remains to ask, $which$ $tree$ do we want at each step? A natural thing is to add the one that optimizes our objective.

Object function:

$$Obj^{(t)} = \sum_{i=1}^n [g_i f_t(x_i) + \frac{1}{2} h_i f_t^2(x_i)] + \Omega(f_t)$$

```python
import xgboost
```

```python
from numpy import loadtxt #numpy函数
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = loadtxt('disease.csv',delimiter=",",skiprows=1)
```

```python
X=dataset[:,0:8]
Y=dataset[:,8]
```

```python
#拆分为训练集和测试集
seed = 7
test_size = 0.33 #测试-训练 数据占比
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=test_size, random_state = seed)

model = XGBClassifier()
eval_set = [(X_test,y_test)]
#early_stopping_rounds = 10: 如果连续10次模型loss值 没有下降就停
model.fit(X_train,y_train,early_stopping_rounds=20,eval_metric='logloss',eval_set=eval_set,verbose=True)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
predictions
```

```python
accuracy = accuracy_score(y_test,predictions)
```
